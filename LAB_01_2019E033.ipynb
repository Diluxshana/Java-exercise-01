```python
#Q1_LAB_01_2019E033

from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
print(data)
```

```python
#Q2_LAB_01_2019E033
import numpy as np

# Shape of the data
print(" *Shape of the data:", data.data.shape)

# Sample count per class
print("\n *Sample count per class:", {n: v for n, v in zip(data.target_names, np.bincount(data.target))})

# Features in the dataset
print("\n *Features in the dataset:", data.feature_names)

# Other properties of the dataset
print("\n *Other properties of the dataset:", data.DESCR)
```

```python
#Q3_LAB_01_2019E033

# Divide the data into X (feature values) and y (target values)
X = data.data
Y = data.target
print(X)
print(Y)
```

```python
#Q4_LAB_01_2019E033
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


# Fit a nearest neighbor model with 5 nearest neighbors
k_nearest_neighbour = KNeighborsClassifier(n_neighbors=5)

# Fit the model on the whole data
k_nearest_neighbour.fit(X, Y)

# Test the model on the same data (no train-test split)
score = k_nearest_neighbour.score(X, Y)
print("Score on the same data:", score)

# Divide the whole data into 80% train and 20% test
x_training_data, x_test_data, y_training_data, y_test_data  = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train the same model (5 nearest neighbors) on the training data
k_nearest_neighbour_training_data = KNeighborsClassifier(n_neighbors=5)
k_nearest_neighbour_training_data.fit(x_training_data, y_training_data)

# Test the trained model on the unseen test data
score_train =k_nearest_neighbour_training_data.score(x_test_data, y_test_data)
print("Score on the test data:", score_train)
```

```python
#training accuracy and testing accuracy
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

neighbors = np.arange(1, 8)
training_accuracy = np.empty(len(neighbors))
testing_accuracy = np.empty(len(neighbors))
print(training_accuracy)
print(testing_accuracy)
  
```

```python
# Loop over K values
for i, k in enumerate(neighbors):
    k_nearest_neighbour = KNeighborsClassifier(n_neighbors=k)
    k_nearest_neighbour.fit(x_training_data, y_training_data)
      
# Compute training and test data accuracy
    training_accuracy[i] = k_nearest_neighbour.score(x_training_data, y_training_data)
    testing_accuracy[i] = k_nearest_neighbour.score(x_test_data, y_test_data)
                                                    
 #plotting graph                                                   
plt.plot(neighbors, testing_accuracy, label = 'Testing dataset Accuracy')
plt.plot(neighbors, training_accuracy, label = 'Training dataset Accuracy')
  
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```
